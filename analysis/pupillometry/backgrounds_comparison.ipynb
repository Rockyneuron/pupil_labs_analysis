### Comparing surprise experiments with gray and black background

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_black=pd.read_csv('black_params.csv')
data_gray=pd.read_csv('gray_params.csv')
```

```python
data_black['back']='black'
data_gray['back']='gray'

data=pd.concat([data_black,data_gray])
data=data.set_index('Unnamed: 0')
data
```

```python
to_compare=data.columns

for pair in to_compare.drop(['type','back']):
    x=data.query("type=='asset'")[pair]
    y= data.query("type=='surprise'")[pair]
    fig,ax=plt.subplots(1)
    toplot=sns.boxplot(data=data,x='type',y=pair,hue='back',boxprops={'alpha': 0.4},palette={'black': 'black', 'gray': 'white'})
    ax.legend(title='background')
    # toplot.set(plt.legend(title='Smoker', loc='upper left')
```

```python

for pair in to_compare.drop(['type','back']):
    x=data.query("type=='asset'")[pair]
    y= data.query("type=='surprise'")[pair]
    fig,ax=plt.subplots(1)
    toplot=sns.boxplot(data=data,x='back',y=pair,hue='type',boxprops={'alpha': 0.4})
    ax.legend(title='background')
    ax.axvspan(-0.5,0.5, color='k', alpha=.3)
    # toplot.set(plt.legend(title='Smoker', loc='upper left')
```
