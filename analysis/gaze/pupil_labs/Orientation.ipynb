```python
import pandas as pd 
import numpy as np
from pathlib import Path
import os
```

```python
recording_location = r"Z:\XSCAPE\Exp_Stgo_0001\ANALIZADOS"
csv_name='ES0001_S029_saccades.csv'
def print_file_structure(startpath):
    for root, dirs, files in os.walk(startpath):
        level = root.replace(startpath, '').count(os.sep)
        indent = ' ' * 4 * (level)
        print(f'{indent}{os.path.basename(root)}/')
        subindent = ' ' * 4 * (level + 1)
        for f in sorted(files):
            print(f'{subindent}{f}')
print_file_structure(recording_location)
```

```python
exported_gaze_csv = os.path.join(recording_location,'ES0001_S029_fixations.csv')
gaze_pd_frame = pd.read_csv(exported_gaze_csv)

gaze_pd_frame.head()
```

```python
gaze_pd_frame['orientation']=gaze_pd_frame['angle']
gaze_pd_frame.head()
```

```python
ori=gaze_pd_frame['angle']
gaze_pd_frame.loc[(ori<0),['orientation']]=gaze_pd_frame.loc[(ori<0),['orientation']]+360
gaze_pd_frame.head()
```

```python
gaze_pd_frame['verticality']=0
gaze_pd_frame.head()
```

```python
vert=gaze_pd_frame['orientation']
gaze_pd_frame.loc[(vert>45) & (vert<135),['verticality']]=1
gaze_pd_frame.loc[(vert>225) & (vert<315),['verticality']]=1                                      
gaze_pd_frame.head()
```
